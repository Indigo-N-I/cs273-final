```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/Users/youli/Desktop/UC_Irvine/WQ_2023/CS273P/project/jigsaw-toxic-comment-classification-challenge-2/train.csv')

# Count the frequency of each category in each column
toxic_count = df['toxic'].value_counts()
severe_toxic_count = df['severe_toxic'].value_counts()
obscene_count = df['obscene'].value_counts()
threat_count = df['threat'].value_counts()
insult_count = df['insult'].value_counts()
identity_hate_count = df['identity_hate'].value_counts()


# Create a bar chart using matplotlib
labels = ['No', 'Yes']
toxic_data = [toxic_count[0], toxic_count[1]]
severe_toxic_data = [severe_toxic_count[0], severe_toxic_count[1]]
obscene_data = [obscene_count[0], obscene_count[1]]
threat_data = [threat_count[0], threat_count[1]]
insult_data = [insult_count[0], insult_count[1]]
identity_hate_data = [identity_hate_count[0], identity_hate_count[1]]

x = [0, 1]
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar([i - width/2 for i in x], toxic_data, width, label='Toxic')
rects2 = ax.bar([i + width/2 for i in x], severe_toxic_data, width, label='Severe Toxic')
rects3 = ax.bar([i + 1.5*width for i in x], obscene_data, width, label='Obscene')
rects4 = ax.bar([i + 2.5*width for i in x], threat_data, width, label='Threat')
rects5 = ax.bar([i + 3.5*width for i in x], insult_data, width, label='Insult')
rects6 = ax.bar([i + 4.5*width for i in x], identity_hate_data, width, label='Identity Hate')

# Add title and axis labels to the plot
ax.set_title('Frequency of Categories')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Frequency')

# Add legend to the plot
ax.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import gensim

sentences = [sentence.split() for sentence in data['comment_text']]
word_embeddings = gensim.models.FastText(sentences=sentences, vector_size=100, window=5, min_count=5, workers=4)

y_train = data[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].values
```

```python

from keras import layers
from keras.models import Sequential


# Prepare the input data
MAX_SEQUENCE_LENGTH = 100

X_train = np.zeros((len(data['comment_text']), MAX_SEQUENCE_LENGTH))
for i, sentence in enumerate(sentences):
    for t, word in enumerate(sentence):
        if t >= MAX_SEQUENCE_LENGTH:
            break
        if word in word_embeddings.wv.key_to_index:
            X_train[i, t] = word_embeddings.wv.key_to_index[word]

# Define the model architecture
model = Sequential()
model.add(layers.Embedding(input_dim=len(word_embeddings.wv.index_to_key), output_dim=100, weights=[word_embeddings.wv.vectors], input_length=MAX_SEQUENCE_LENGTH, trainable=False))
model.add(layers.Conv1D(filters=128, kernel_size=5, activation='relu'))
model.add(layers.GlobalMaxPooling1D())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(6, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)


```
